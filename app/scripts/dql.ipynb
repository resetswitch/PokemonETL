```python
import os
from dotenv import load_dotenv
from db_schema import get_ddb_conn
load_dotenv()

db_name = os.getenv("POSTGRES_DB")
conn = get_ddb_conn(db_name=db_name)
```

```python
conn.execute("""
    SELECT table_schema,
           table_name,
           column_name,
           ordinal_position,
           data_type,
           is_nullable,
           column_default
    FROM information_schema.columns
    WHERE table_schema NOT IN ('information_schema','pg_catalog')
    ORDER BY table_schema, table_name, ordinal_position
""").df()
```
